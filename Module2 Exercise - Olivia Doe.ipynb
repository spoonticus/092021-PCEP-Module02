#### 1. Find the average score
A program is needed to prompt the user operator for exam scores, calculate the average score and display the result to the screen. The average score is calculated as the sum of the scores divided by the number of scores.

##### Solution:



   **INPUT:** Exam scores
   
   
   **PROCESS:**
   
           BEGIN
           
                CREATE sum and counter
                
                INITIALIZE sum = 0
                
                INITIALIZE counter = 1
                
                WHILE ask if user want to input an exam score
                
                    PRINT ask user to input an exam score
                    
                    GET input
                    
                    ADD input to sum
                    
                    INCREMENT counter by 1
                    
                ENDWHILE
                
                PRINT sum divides by counter 
               
           END
           
   **OUTPUT:** Average score
 

#### 2. How much water runs off a roof in a rain storm?
To calculate the runoff from any given rainfall: Take the dimensions of the footprint of the roof and convert them to inches. (So, a 50' x 20' roof is 600" x 240"). Multiply the roof dimensions by the number of inches of rainfall. As an example, 600" x 240" x 1" = 144,000 cubic inches of water for an inch of rainfall. Divide that result by 231 to get the number of gallons (because 1 gallon = 231 cubic inches). (144,000/231 = 623.38).

##### Solution:


    **INPUT:** Dimensions of the footprint of the roof and the number of inches of rainfall
    
    **PROCESS:**
    
    
                BEGIN
                
                    CREATE length, width, rainfall, runoff
                    
                    INITIALIZE length = 0, 
                    
                               width = 0,
                               
                               rainfall = 0,
                               
                               runoff = 0
                    
                    
                    PRINT "Enter the length of the roof in feet: "
                    
                    INPUT to length
                    
                    PRINT "Enter the width of the roof in feet: "
                    
                    INPUT to width
                    
                    PRINT "Enter number of inches of rainfall: "
                    
                    INPUT to rainfall
                    
                    runoff = ((length / 12) * (width / 12) * rainfall) / 231
                    
                    PRINT "The amount of expected water runs off is " + runoff + " gallons."
                    
                END
                    
    
    
    
    **OUTPUT:** Amount of water runs off a roof in a rain storm in gallons

#### 3. Process customer record
A program is required to read a customer’s name, a purchase amount and a tax code. The tax code has been validated and will be one of the following: 0 tax exempt (0%) 1 state sales tax only (3%) 2 federal and state sales tax (5%) 3 special sales tax (7%) The program must then compute the sales tax and the total amount due,and print the customer’s name, purchase amount, sales tax and total amount due.

**INPUT:** customer's name, purchase amount, tax code

**PROCESS:**

        BEGIN
        
            CREATE name, purchaseAmount, taxCode, taxPercent, taxAmount, totalAmount
            
            INITIALIZE purchaseAmount = 0,
            
                        taxCode = -1,
                        
                        taxPercent = -1,
                        
                        taxAmount = 0,
                        
                        totalAmount = 0
                        
            PRINT "Enter name: "
            
            INPUT to name
            
            PRINT "Enter purchase amount: "
            
            INPUT to purchaseAmount
            
            PRINT "Pick a tax code: 
            
                    0. tax exempt (0%)
                    1. state sales tax only (3%)
                    2. federal and state sales tax (5%)
                    3. special sales tax (7%)"
                    
            INPUT to taxCode
            
            SWITCH (taxCode)
            
                0:
                    taxPercent = 0
                    
                1:
                    taxPercent = 0.03
                    
                2:
                    taxPercent = 0.05
                    
                3:
                    taxPercent = 0.07
            
            ENDSWITCH
            
            taxAmount = purchaseAmount * taxPercent
            
            totalAmount = taxAmount + purchaseAmount
            
            PRINT "Customer's name: " + name
            
            PRINT "Purchase Amount: '$' "  + purchaseAmount
            
            PRINT "Sales Tax: \\$" + taxAmount
            
            PRINT "Total Amount: \\$" + totalAmount
            
        END
            
            
                   

**OUTPUT:** customer's name, purchase amount, sales tax, total amount

#### 4. Calculate employee's pay 
A program is required by a company to read an employee’s number, pay rate and the number of hours worked in a week. The program is then to validate the pay rate field and the hours worked field and, if valid, compute the employee’s weekly pay and then print it and the input data.

Validation: According to the company’s rules, the maximum hours an employee can work per week is 60 hours, and the maximum hourly rate is $25.00 per hour. If the hours worked field or the hourly rate field is out of range, the input data and an appropriate message are to be printed and the employee’s weekly pay is not to be calculated.

Weekly pay calculation: Weekly pay is calculated as hours worked times pay rate. If more than 35 hours are worked, payment for the overtime hours worked is calculated at time-and-a-half.

##### Solution: 

**INPUT:** employee's number, pay rate, number of hours worked in a week

**PROCESS:** 

            BEGIN
            
                    CREATE employeeNumber, payRate, hoursWorked, weeklyPay
                    
                    INITIALIZE employeeNumber = 0,
                                payRate = 0,
                                hoursWorked = 0,
                                weeklyPay = 0
                                
                    PRINT "Enter Employee's Number: "
                    
                    INPUT to employeeNumber
                    
                    PRINT "Enter Employee's Pay Rate: "
                    
                    INPUT to payRate
                    
                    IF payRate > 25
                    
                        PRINT "Employee's pay rate cannot be higher than $25.00 per hour" 
                        
                        PRINT "Please contact your HR representative"
                        
                        EXIT error code
                        
                    ENDIF
                    
                    PRINT "Enter Number of Hours Worked in a Week: "
                    
                    INPUT to hoursWorked
                    
                    IF hoursWorked > 60
                        
                        PRINT "Employee cannot work more than 60 hours a week."
                        
                        PRINT "Please contact your HR representative"
                        
                        EXIT error code
                        
                    ENDIF
                    
                    
                    IF hoursWorked > 35
                        
                        weeklyPay = (35 * payRate) + ( (hoursWorked - 35) * (payRate * 1.5))
                        
                    ELSE
                    
                        weeklyPay = hoursWorked * payRate
                        
                    ENDIF
                    
                    
                    PRINT "Employee " + employeeNumber + "'s weekly pay: " 
                    
                    PRINT weeklyPay           
            
            END


**OUTPUT:** if valid: weekly pay 
            if no valid: error message
